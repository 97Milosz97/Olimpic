## GDP, Population and Politics matter?

```python
olympics
```

```python
medals_per_country = pd.crosstab(index = olympics["Country"], columns=olympics["Medal"], margins = True, margins_name="Total")
medals_per_country # the frequency of medals
```

```python
medals_per_country.drop("Total", axis = 0, inplace = True)
#Deliting last row "Total"
```

```python
medals_per_country = medals_per_country[["Total", "Gold", "Silver", "Bronze"]] 
#Change columns
medals_per_country.sort_values(by = ["Total", "Gold", "Silver", "Bronze"] ,ascending = False, inplace = True)
```

```python
medals_per_country = medals_per_country.merge(dic, how = "left", left_index = True,
                                             right_on = "Country")
# Merge dic df
```

```python
medals_per_country=  medals_per_country.drop(columns=["Code"]).set_index("Country") 
# Droping Code column and set index to Country
```

```python
olympics["Games"] = olympics.apply(lambda x : str(x["Year"]) + " " + x["City"], axis = 1)
# Adding new coolumn with Year and City
```

```python
olympics.groupby("Country").apply(lambda x: x["Games"].nunique()) 
# Unique Games by each country
```

```python
medals_per_country["Total Games"] = olympics.groupby("Country").apply(lambda x: x["Games"].nunique()) 
#Adding new column Unique Games by each country
```

```python
medals_per_country.rank(ascending=False) #Seting rank 
```

## Statisitcal Analysis and Hypothetis Testing with scipy

```python
medals_per_country.drop(columns=["Gold", "Silver", "Bronze"], inplace = True) #droping columns
```

```python
medals_per_country.corr(method="pearson") # Normal destribution
```

```python
medals_per_country.corr(method="spearman") # Not Normal desctrubution, relative rank
```

## Hypothesis 1: There is no relationship between Total Medals and Population

```python
stats.spearmanr(medals_per_country["Total"], medals_per_country["Population"], nan_policy="omit")
# moderate correlation, pi value below 0.05 - statistically significant correlation Reject the hypothesis
```

## Hypothesis 2: There is no relationship between Total Medals and GDP per Capita

```python
stats.spearmanr(medals_per_country["Total"], medals_per_country["GDP per Capita"], nan_policy="omit")
# moderate correlation, pi value below 0.05 - statistically significant correlation Reject the hypothesis
```

## Hypothesis 3: There is no relationship between Total Medals and Prticipations

```python
stats.spearmanr(medals_per_country.head(50)["Total"], medals_per_country.head(50)["Total Games"], nan_policy="omit")
# correlation, pi value below 0.05 - statistically significant correlation Reject the hypothesis
```

## Summer vs Winter 

```python
olympics
```

```python
sum_vs_win = pd.crosstab(olympics["Country"], olympics["Edition"]) # creatign pivot table
```

```python
sum_vs_win
```

```python
men_vs_women = pd.crosstab(olympics["Country"], olympics["Gender"]) # creatign pivot table
```

```python
medals_by_category = pd.concat([sum_vs_win, men_vs_women], axis = 1) # creatign contact data frame
```

```python
medals_by_category["Total"] = medals_by_category["Summer"] + medals_by_category["Winter"] # creating new column
```

```python
medals_by_category.sort_values(by="Total", ascending= False, inplace = True) # Sorting by Total
```

```python
medals_by_category
```

```python
ranks = medals_by_category.rank(ascending=False, method = "average") # Creating rank dataframe
```

```python
top_50 = ranks.head(50)
```

```python
plt.figure(figsize= (50,10))
sns.heatmap(top_50.T,cmap="RdYlGn_r", annot = True, fmt='2.0f')
plt.show()
# Creating heatmap
```

## Summer Games vs. Winter Games - does Geographical Loation matter

```python
plt.figure(figsize=(50,10))
sns.heatmap(top_50.sort_values("Winter").T,cmap="RdYlGn_r", annot = True, fmt='2.0f')
plt.show()
```
